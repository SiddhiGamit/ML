```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('airline-passenger-traffic.csv', header = None)
```

```python
df.head()
```

```python
df.columns = ['Months', 'Passengers']
```

```python
df
```

```python
df.info()
```

```python
df['Months'] = pd.to_datetime(df['Months'], format = '%Y-%m')
```

```python
df
```

```python
df.info()
```

```python
df = df.set_index('Months')
```

```python
df
```

```python
df.plot(figsize=(12,4))
plt.legend(loc = 'best')
plt.title('Airline Passenger Traffic')
plt.show()
```

## Time Series - Data Cleaning

### Impute the Missing Values

```python
## 1. Mean Imputation
df['Passengers'].fillna(df['Passengers'].mean()).plot(figsize = (12,4))
plt.legend(loc = 'best')
plt.title('Airline Passenger Traffic')
plt.show()
```

```python
## 2 Last Observation Carry Forward
df['Passengers'].fillna(method = 'bfill').plot(figsize = (12,4))
plt.legend(loc = 'best')
plt.title('Airline Passenger Traffic')
plt.show()
```

```python
df.plot(figsize = (12,4))
plt.legend(loc = 'best')
plt.title('Airline Passenger Traffic')
plt.show()
```

```python
# Linear Inter Polation
df['Passengers'] = df['Passengers'].interpolate(method = 'linear')
```

```python
## 3 Linear Interpolation
df.plot(figsize = (12,4))
plt.legend(loc = 'best')
plt.title('Airline Passenger Traffic')
plt.show()
```

## Smoothing technique

##dataset-->split-->train and test

```python
df.info()
```

```python
train=df[:120]
test=df[120:]
```

```python
train.plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Airline Passengers Traffic')
plt.show()
```

```python
test.plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Airline Passengers Traffic')
plt.show()
```

## 1 Naive Technique 


- Last observation carried forward

```python
train.iloc[119][0]
```

```python
test['Naive']=train.iloc[119][0]
```

```python
test.head()
```

```python
train.plot(figsize=(12,4))
test['Passengers'].plot(figsize=(12,4))
test['Naive'].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Airline Passengers Traffic')
plt.show()
```

```python
train['Passengers'].mean()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(test['Passengers'], test['Naive'])).round(2)
```

```python
rmse
```

```python
mape = np.round(np.mean(np.abs(test['Passengers']-test['Naive'])/test['Passengers'])*100,2)
```

```python
mape
```

## 2. Simple Avg

```python
test['Simple Avg']=train['Passengers'].mean()
```

```python
train.plot(figsize=(12,4))
test['Passengers'].plot(figsize=(12,4))
test['Simple Avg'].plot(figsize=(12,4))
plt.legend(loc='best')
plt.title('Airline Passengers Traffic')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(test['Passengers'], test['Simple Avg'])).round(2)
```
